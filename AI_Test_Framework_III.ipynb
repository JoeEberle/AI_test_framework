```python
def clean_string(input_string):
    input_string = input_string.replace("‐", " ")  # replace minus signs with blank
    input_string = input_string.replace("﴾", "(")  # squiggle parenthesis with parenthesis 
    input_string = input_string.replace("﴿", ")")  # squiggle parenthesis with parenthesis     

    unwanted_chars = '"\'‐-–:/“”‘’'  # Add any other quote-like or minus/dash characters
    translation_table = str.maketrans('', '', unwanted_chars)

    cleaned_string = input_string.translate(translation_table)
    return cleaned_string


def outmd(definition):
    definition = clean_string(definition) 
    with open(file_name, 'a', encoding='utf-8') as f:
        f.write(definition)  
    display(Markdown(definition))
```

```python
import re

#     "healthcare": {"cancer", "diabetes", "glucose", "A1C", "surgery","blood", "screening" , "lungs", "bone", "colonoscopy",
#                    "tissue", "procedure", "doctor", "heart", "disease", "diagnosis", "population", "patient", "hospital"
#                    , "medication"}


# Define keyword lists
CATEGORY_KEYWORDS = {
    "penguin": {"flipper", "beak", "gentoo", "penguins", "feathers", "colony", "antarctica", "huddle"},
    "titanic": {"passenger", "ticket class", "survived", "died", "titanic", "iceberg", "lifeboat", "drown", "ship", "deck"},
    "healthcare": {"cancer", "diabetes", "glucose", "A1C", "surgery","blood", "screening" , "lungs", "bone", "colonoscopy",
                   "tissue", "procedure", "doctor", "heart", "disease", "diagnosis", "population", "patient", "hospital"
                   ,"condition", "sick", "ill", "visit", "clinic", "nurse", "smoking", "asthma", "covid"
                   ,"dr", "practice", "diet", "drug", "treat", "treatment", "walking", "weight", "obese", "health"   
                   ,"pnuemonia", "fever", "rash", "infarction", "myocardial", "neuropathy", "sepsis", "wellness"
                   ,"knee", "arm", "hip", "EKG", "leg", "shoulder", "mouth", "eye", "teeth", "biopsy", "lab"
                   , "epidemic", "pandemic"  
                   ,"depression", "anxiety", "abuse", "overweight", "prevalance", "treatment", "walking", "weight", "obese", "health"                     
                   ,"chronic", "diabetic", "at risk", "viagra", "ozempic", "warfarin", "wagovi", "prozac", "opiod", "addiction"  
                   ,"doctor", "stroke", "medical", "prescription", "tumor", "comorbid", "cough", "congestion", "xray", "scan"                        
                   , "medication", "prescribe", "medicine", "skin"},   
    "innapropriatelanguage": {"shit", "fuck", "crap", "sucks", "bitch", "asshole", "pussy", "dick" , "cunt",  "bullshit",},
    "hostility": {"bomb", "kill", "shoot", "invade", "weapon", "gun", "hate", "bully"}, 
    "racism": {"black people", "nigger", "jew", "jews", "white people", "black people"}    
}

def keyword_based_classifier(text):
    """
    Classifies a given text into 'penguin', 'titanic', or 'healthcare' based on keyword presence.
    Returns the category with the most keyword matches or 'Unknown' if no match is found.
    """
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    scores = {category: 0 for category in CATEGORY_KEYWORDS}

    for category, keywords in CATEGORY_KEYWORDS.items():
        for keyword in keywords:
            if re.search(rf"\b{re.escape(keyword)}\b", text):  # Match whole words only
                scores[category] += 1

    # Find category with highest score
    best_match = max(scores, key=scores.get)
    
    return best_match if scores[best_match] > 0 else "Unknown"

# Example Usage
sentences = [
    "The gentoo penguin has a distinctive beak and flipper movement.",
    "The Titanic had over 1,200 passengers and hit an iceberg.",
    "Early diagnosis of heart disease can save lives.",    
    "I had a cancer screening.",    
    "I had a colonoscopy.",
    "I love tacos.",    
    "The rocket launch was delayed due to weather conditions."
]

for sentence in sentences:
    print(f"Text: {sentence}\nCategory: {keyword_based_classifier(sentence)}\n")
```

```python
from textblob import TextBlob
def detect_toxicity(text):
    acceptable_toxicity_level = -0.3
    sentiment = TextBlob(text).sentiment.polarity
    return "Toxic" if sentiment < acceptable_toxicity_level else "Neutral" , sentiment, text
```

```python
import re

#     "healthcare": {"cancer", "diabetes", "glucose", "A1C", "surgery","blood", "screening" , "lungs", "bone", "colonoscopy",
#                    "tissue", "procedure", "doctor", "heart", "disease", "diagnosis", "population", "patient", "hospital"
#                    , "medication"}


# Define keyword lists
CATEGORY_KEYWORDS = {
    "penguin": {"flipper", "beak", "gentoo", "penguins", "feathers", "colony", "antarctica", "huddle", "penguins"},
    "titanic": {"passenger", "ticket class", "survived", "died", "titanic", "iceberg", "lifeboat", "drown", "ship", "deck"},
    "healthcare": {"cancer", "diabetes", "glucose", "A1C", "surgery","blood", "screening" , "lungs", "bone", "colonoscopy",
                   "tissue", "procedure", "doctor", "heart", "disease", "diagnosis", "population", "patient", "hospital"
                   ,"condition", "sick", "ill", "visit", "clinic", "nurse", "smoking", "asthma", "covid"
                   ,"dr", "practice", "diet", "drug", "treat", "treatment", "walking", "weight", "obese", "health"   
                   ,"pnuemonia", "fever", "rash", "infarction", "myocardial", "neuropathy", "sepsis", "wellness"
                   ,"knee", "arm", "hip", "EKG", "leg", "shoulder", "mouth", "eye", "teeth", "biopsy", "lab"
                   , "epidemic", "pandemic"  
                   ,"depression", "anxiety", "abuse", "overweight", "prevalance", "treatment", "walking", "weight", "obese", "health"                     
                   ,"chronic", "diabetic", "at risk", "viagra", "ozempic", "warfarin", "wagovi", "prozac", "opiod", "addiction"  
                   ,"doctor", "stroke", "medical", "prescription", "tumor", "comorbid", "cough", "congestion", "xray", "scan"                        
                   , "medication", "prescribe", "medicine", "skin"},   
    "innapropriatelanguage": {"shit", "fuck", "crap", "sucks", "bitch", "asshole", "pussy", "dick" , "cunt",  "bullshit",},
    "hostility": {"bomb", "kill", "shoot", "invade", "weapon", "gun", "hate", "bully"}, 
    "racism": {"black people", "nigger", "jew", "jews", "white people", "black people"}    
}

SUB_DOMAIN_KEYWORDS = {
    "penguin": {"flipper", "beak", "gentoo", "penguins", "feathers", "colony", "antarctica", "huddle"},
    "titanic": {"passenger", "ticket class", "survived", "died", "titanic", "iceberg", "lifeboat", "drown", "ship", "deck"},
    "disease": {"cancer", "diabetes", "obesity", "myocardial","asthma","cholesterol","blood pressure","influenza","flu","covid","illness","pnuemonia"},
    "condition": {"pregnancy"},    
    "weight": {"obese", "fat", "skinny", "diet","weight loss","overweight"},
    "medication": {"ozempic", "wagovy", "aspirin", "viagra","opiod","statin","anti-depressent","anti-depressent","drug"},                
    "lab result": {"screening", "test", "panel", "ekg","mri","xray","biopsy","screening","colonoscopy"},    
    "anatomy": {"lungs", "bone", "heart"}, 
    "behavioral": {"depression", "anxietty", "addiction", "suicide","abuse"}, 
    "medical": {"doctor", "nurse", "appointment", "visit","opiod","statin","anti-depressent","anti-depressent","pills","drug"},  
    "procedure": {"surgery", "operation", "excission", "ablation","knee replacement","hip replacement"} 
}

def keyword_based_classifier(text):
    """
    Classifies a given text into 'penguin', 'titanic', or 'healthcare' based on keyword presence.
    Returns the category with the most keyword matches or 'Unknown' if no match is found.
    """
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    scores = {category: 0 for category in CATEGORY_KEYWORDS}

    for category, keywords in CATEGORY_KEYWORDS.items():
        for keyword in keywords:
            if re.search(rf"\b{re.escape(keyword)}\b", text):  # Match whole words only
                scores[category] += 1

    # Find category with highest score
    best_match = max(scores, key=scores.get)
    
    return best_match if scores[best_match] > 0 else "Unknown"

def sub_domain_classifier(text):
    """
    Classifies a given text into 'penguin', 'titanic', or 'healthcare' based on keyword presence.
    Returns the category with the most keyword matches or 'Unknown' if no match is found.
    """
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    scores = {category: 0 for category in SUB_DOMAIN_KEYWORDS}

    for category, keywords in SUB_DOMAIN_KEYWORDS.items():
        for keyword in keywords:
            if re.search(rf"\b{re.escape(keyword)}\b", text):  # Match whole words only
                scores[category] += 1

    # Find category with highest score
    best_match = max(scores, key=scores.get)
    
    return best_match if scores[best_match] > 0 else "Unknown"
    

# Example Usage
sentences = [
    "The gentoo penguin has a distinctive beak and flipper movement.",
    "The Titanic had over 1,200 passengers and hit an iceberg.",
    "Early diagnosis of heart disease can save lives.",    
    "I had a cancer screening.",    
    "I had a colonoscopy.",
    "I love tacos.",    
    "The rocket launch was delayed due to weather conditions."
]

for sentence in sentences:
    print(f"Text: {sentence}\nCategory: {keyword_based_classifier(sentence)}\n")
```
